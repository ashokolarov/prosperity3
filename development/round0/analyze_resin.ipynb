```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from log_processing import process_log, from_csv
```

```python
round = 1
day = 2
activities, trades = from_csv(round, day, dir="../data")
```

```python
activities["product"].unique()
```

```python
PRODUCT = "RAINFOREST_RESIN"
activities = activities[activities["product"] == PRODUCT].reset_index()
trades = trades[trades["symbol"] == PRODUCT].reset_index()
```

```python
bb = activities["bid_price_1"]
ba = activities["ask_price_1"]

fig, axs = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
axs[0].hist(bb, bins=100, alpha=0.5, label="bid")
axs[0].set_title("Best Bid Distribution")
axs[0].set_xlabel("Price")
axs[0].grid()

axs[1].hist(ba, bins=100, alpha=0.5, label="ask")
axs[1].set_title("Best Ask Distribution")
axs[1].set_xlabel("Price")
axs[1].grid()

for ax in axs:
    ax.ticklabel_format(useOffset=False, style='plain')

plt.tight_layout()
plt.savefig(f"../research/round0/rr_price_distribution_{round}{day}.pdf")
```
