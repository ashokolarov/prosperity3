```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/r0/r0_notrades.csv", sep=";")
t = df["timestamp"]
```

```python
resin_price = df["mid_price"].where(df["product"] == "RAINFOREST_RESIN").ffill()
resin_max = resin_price.max()
resin_min = resin_price.min()
resin_mean = resin_price.mean()
resin_std = resin_price.std()

print("\033[1mResin Price Info\033[0m")
print(f"Max price: {resin_max}")
print(f"Min price: {resin_min}")
print(f"Mean price: {resin_mean:.1f}")
print(f"Standard deviation: {resin_std:.1f}")

kelp_price = df["mid_price"].where(df["product"] == "KELP").ffill()

fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

axs[0].plot(t, resin_price)
axs[0].axhline(resin_max, label=f'Max price of {resin_max}', c='red')
axs[0].axhline(resin_min, label=f'Min price of {resin_min}', c='orange')
axs[0].grid()
axs[0].legend(loc='upper center', ncol=2, bbox_to_anchor=(0.5, 1.1))


axs[1].plot(t, kelp_price)
axs[1].grid()
axs[1].set_xlabel("Iteration")

plt.tight_layout()
plt.show()
```

```python
resin_bb = df["bid_price_1"].where(df["product"] == "RAINFOREST_RESIN").ffill() - 1E4
resin_ba = df["ask_price_1"].where(df["product"] == "RAINFOREST_RESIN").ffill() - 1E4 

fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].hist(resin_bb, bins=50, alpha=0.5)
axs[0].set_title("Best Bid Price")
axs[0].grid()

axs[1].hist(resin_ba, bins=50, alpha=0.5)
axs[1].set_title("Best Ask Price")
axs[1].grid()
plt.plot()
```
